```
import re, datetime
from dateutil.parser import parse

s = "I have a meeting on 2018-12-10 in New York. mkd 2011-01-04"
match = re.findall('\d{4}-\d{2}-\d{2}', s)
res = []
for m in match:
  a = parse(m, fuzzy_with_tokens=True)
  res.append(a[0].strftime("%Y")+a[0].strftime("%m")+a[0].strftime("%d"))
# date = datetime.datetime.strptime(match.group(), '%Y-%m-%d').date()
# print(date)


from dateparser.search import search_dates
search_dates("find 12/15/18 in this string")
```
